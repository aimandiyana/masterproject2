```python
import pandas as pd

# Classification report metrics
classification_report_data = {
    'Precision': [0.99, 0.03],
    'Recall': [1.00, 0.03],
    'F1-Score': [1.00, 0.03],
    'Support': [552824, 2895]
}

# Other metrics
roc_auc_score = 0.5103368405506765
accuracy_score = 0.9900615238996687
time_taken = 14790.54471373558

# Create a DataFrame
df_classification_report = pd.DataFrame(classification_report_data, index=['0', '1'])
df_metrics = pd.DataFrame({
    'ROC AUC Score': [roc_auc_score],
    'Accuracy Score': [accuracy_score],
    'Time Taken (seconds)': [time_taken]
})

# Save DataFrames to an Excel file
with pd.ExcelWriter('svm_classification_metrics.xlsx') as writer:
    df_classification_report.to_excel(writer, sheet_name='Classification Report')
    df_metrics.to_excel(writer, sheet_name='Metrics')
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Replace the values in your confusion matrix
conf_matrix = [[552547, 277],
               [869, 2026]]

# Define class labels
class_labels = ['Non-Fraud', 'Fraud']

# Create confusion matrix plot
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix for Conventional RF Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
